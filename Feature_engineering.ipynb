
1. What is a parameter?

-> A parameter is a model component that is learned from the training data. For example, in linear regression, the slope and intercept are parameters.

2. What is correlation?  

->Correlation is a measure of the relationship between two variables. It ranges from -1 to +1 and shows how one variable changes with respect to another.

3. What does negative correlation mean?

->Negative correlation means that as one variable increases, the other decreases. For example, as temperature increases, the need for blankets decreases.

4. Define Machine Learning. What are the main components in Machine Learning?  

->Machine Learning is a field of AI where systems learn patterns from data and make predictions or decisions. Its main components are data, model, loss function, optimizer, and evaluation.

5. How does loss value help in determining whether the model is good or not?  

->The loss value indicates how well the model is performing. A low loss means the model's predictions are close to actual values.
